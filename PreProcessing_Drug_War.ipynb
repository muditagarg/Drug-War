Cleaning the tweet text

```python
import pandas as pd
df = pd.read_csv(r"/content/Ouput.csv")

df
```

Dropping columns which are not needed

```python
df= df.drop(columns= 'Unnamed: 10')
```

```python
df= df.drop(columns= 'Unnamed: 11')
```

```python
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
```

Text Cleaning

```python
import re
# Reading emoji patterns which contains emoticons , symbols & pictographs, transport & map symbols , flags
emojisCollection = re.compile("["
                       u"\U0001F600-\U0001F64F"  
                       u"\U0001F300-\U0001F5FF"  
                       u"\U0001F680-\U0001F6FF"  
                       u"\U0001F1E0-\U0001F1FF"  
                       u"\U00002702-\U000027B0"
                       u"\U000024C2-\U0001F251"
                       "]+", flags=re.UNICODE)
df['CleanedTweet'] = (df['text'].str.lower().str.replace(r'[^\w\s]+', '').str.replace(emojisCollection, '').str.strip())
```

Saving it into a new data frame and exporting the cleaned file
